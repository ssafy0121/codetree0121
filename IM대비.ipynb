1번 칸부터 N번째 칸까지 순서대로 총 N개의 칸이 있습니다. 이 중 A
i

 번째 칸부터 B

​
 번째 칸까지 각각 블럭을 1씩 쌓으라는 명령이 총 K번 주어집니다. (1≤i≤K)

명령을 다 수행한 이후 1번 칸부터 N번 칸까지 쌓인 블럭의 수 중 최댓값을 출력하는 프로그램을 작성해보세요.

```python
n, k = map(int, input().split())
commands = [tuple(map(int, input().split())) for _ in range(k)]

# Please write your code here.

# 1. 블럭을 쌓을 n개의 칸 생성 (인덱스 편의를 위해 n+1 크기 사용)
# 처음에는 블럭이 전혀 놓여 있지 않으므로 0으로 초기화합니다.
blocks = [0] * (n + 1)

# 2. k번의 명령을 하나씩 수행
for a, b in commands:
    # a번째 칸부터 b번째 칸까지 블럭을 1씩 쌓습니다.
    for i in range(a, b + 1):
        blocks[i] += 1

# 3. 모든 명령 수행 후, 쌓인 블럭의 수 중 최댓값 찾기
# 내장함수 max() 사용이 금지될 수 있으므로 직접 비교 로직을 사용합니다.
max_blocks = 0
for count in blocks:
    if count > max_blocks:
        max_blocks = count

# 4. 최댓값 출력
print(max_blocks)
```

1차원 직선 상에 N개의 선분이 놓여 있습니다. 가장 많이 겹치는 구간에서는, 몇 개의 선분이 겹치는지를 구하는 프로그램을 작성해보세요. 단, 겹치는 구간을 찾는 문제이므로 끝점에서 닿는 경우는 겹치는 것으로 생각하지 않습니다.

```python
# 1. 입력 받기
n = int(input())
segments = [tuple(map(int, input().split())) for _ in range(n)]

# 2. 좌표를 기록할 배열 생성 (범위: -100 ~ 100이므로 200칸 설정)
# offset을 100으로 두어 음수 좌표를 양수로 변환합니다.
OFFSET = 100
checked = [0] * 201

# 3. 각 선분이 지나는 구간 표시하기
for x1, x2 in segments:
    # x1부터 x2-1까지의 구간에 +1을 해줍니다.
    # 이렇게 하면 x2(끝점)에서 만나는 경우는 카운트되지 않습니다.
    for i in range(x1 + OFFSET, x2 + OFFSET):
        checked[i] += 1

# 4. 최댓값 찾기 및 출력
max_overlap = max(checked)
print(max_overlap)
```

위치 0에서 시작하여 N번의 명령에 걸쳐 움직인 뒤, 2번 이상 지나간 영역의 크기를 출력하는 프로그램을 작성해보세요. 단 명령은 “x L“, “x R” 형태로만 주어집니다. "x L" 의 경우 왼쪽으로 x만큼 이동해야 함을, "x R"의 경우 오른쪽으로 x만큼 이동해야 함을 뜻합니다.

```python
# 1. 입력 받기
n = int(input())

# 2. 필요한 변수들 준비
current_pos = 0  # 로봇(혹은 사람)의 현재 위치 (0에서 시작)
OFFSET = 1000    # 음수 좌표를 처리하기 위한 여유 공간
checked = [0] * 2001 # 이동 경로를 기록할 배열 (전체 칸)

# 3. N개의 명령 처리하기
for _ in range(n):
    line = input().split()
    distance = int(line[0]) # 이동 거리 x
    direction = line[1]    # 방향 L 또는 R

    if direction == 'R':
        # 오른쪽 이동: 현재 위치에서 (현재 + 거리)까지 칸을 채움
        start_pos = current_pos
        end_pos = current_pos + distance

        for i in range(start_pos + OFFSET, end_pos + OFFSET):
            checked[i] += 1

        current_pos = end_pos # 이동 후 위치 갱신

    else: # direction == 'L'
        # 왼쪽 이동: (현재 - 거리)부터 현재 위치까지 칸을 채움
        start_pos = current_pos - distance
        end_pos = current_pos

        for i in range(start_pos + OFFSET, end_pos + OFFSET):
            checked[i] += 1

        current_pos = start_pos # 이동 후 위치 갱신

# 4. 2번 이상 지나간 영역의 크기(칸의 개수) 계산
overlap_count = 0
for count in checked:
    if count >= 2: # 2번 이상 지나갔다면
        overlap_count += 1

print(overlap_count)
```

일직선으로 무한히 나열된 타일이 있습니다. 아무 타일에서 시작하여 N번의 명령에 걸쳐 움직입니다.

명령은 x L, x R 형태로만 주어지며, x L의 경우 왼쪽으로 이동하면서 현재 위치 타일포함 총 x칸의 타일을 흰색으로 연속하게 칠하고, x R의 경우 오른쪽으로 이동하면서 현재 위치 타일포함 총 x칸의 타일을 검은색으로 연속하게 칠함을 뜻합니다.

각 명령 이후에는 마지막으로 칠한 타일 위치에 서있는다고 가정합니다. 타일의 색은 덧칠해지면 마지막으로 칠해진 색으로 바뀌는데, 만약 타일 하나가 순서 상관없이 흰색과 검은색으로 각각 두 번 이상 칠해지면 회색으로 바뀌고 더 이상 바뀌지 않습니다.

모든 명령을 실행한 뒤의 흰색, 검은색, 회색의 타일 수를 각각 출력하는 프로그램을 작성해보세요.

```python
# 1. 입력 받기
# n: 명령의 개수입니다.
n = int(input())

# 2. 필요한 설정값 준비
# OFFSET(오프셋): 좌표가 마이너스로 가는 것을 막기 위해 더해주는 '기준점'입니다.
# n이 최대 1000이고 x가 100이므로, 최대 100,000까지 움직일 수 있습니다.
# 넉넉하게 10만이라는 값을 기준으로 잡습니다.
OFFSET = 100000
MAX_SIZE = 200001 # 왼쪽/오른쪽 모두 고려한 전체 배열 크기입니다.

# 타일의 정보를 저장할 리스트(배열)들을 만듭니다.
# 'count_white' (카운트 화이트): 각 타일이 흰색으로 몇 번 칠해졌는지 기록
count_white = [0] * MAX_SIZE
# 'count_black' (카운트 블랙): 각 타일이 검은색으로 몇 번 칠해졌는지 기록
count_black = [0] * MAX_SIZE
# 'tile_color' (타일 컬러): 현재 타일의 최종 색깔 (0:없음, 1:흰색, 2:검은색, 3:회색)
tile_color = [0] * MAX_SIZE

# 로봇의 현재 위치를 저장하는 변수입니다. 0번지(기준점)에서 시작합니다.
current_pos = 0

# 3. N개의 명령을 하나씩 처리합니다.
for _ in range(n):
    # 입력 예시 "4 R"을 받아서 거리(4)와 방향(R)으로 나눕니다.
    temp_input = input().split()
    distance = int(temp_input[0]) # 거리 x
    direction = temp_input[1]    # 방향 (L 또는 R)

    if direction == 'R': # 오른쪽으로 갈 때 (검은색 칠하기)
        # 현재 위치 타일 포함 총 x칸이므로, 반복 범위를 정합니다.
        # start(시작)부터 end(끝)까지 x칸을 이동합니다.
        start = current_pos
        end = current_pos + distance

        for i in range(start, end):
            # 이 타일이 이미 회색(3번)이라면 아무것도 하지 않고 넘어갑니다.
            if tile_color[i + OFFSET] == 3:
                continue

            # 검은색 칠하기 횟수를 1 늘립니다.
            count_black[i + OFFSET] += 1

            # 만약 흰색과 검은색이 둘 다 2번 이상 칠해졌다면 회색(3번)으로 바뀝니다.
            if count_white[i + OFFSET] >= 2 and count_black[i + OFFSET] >= 2:
                tile_color[i + OFFSET] = 3
            else:
                # 그렇지 않다면 일단 현재 색깔은 검은색(2번)입니다.
                tile_color[i + OFFSET] = 2

        # 마지막으로 칠한 타일 위치에 서 있게 됩니다.
        # (x칸 이동했으므로 위치는 end - 1이 됩니다.)
        current_pos = end - 1

    else: # direction == 'L' 일 때 (왼쪽으로 갈 때, 흰색 칠하기)
        # 왼쪽으로 이동하므로 시작점은 현재 위치에서 거리를 뺀 곳입니다.
        start = current_pos - distance + 1
        end = current_pos + 1

        for i in range(start, end):
            # 이미 회색(3번)이면 패스!
            if tile_color[i + OFFSET] == 3:
                continue

            # 흰색 칠하기 횟수를 1 늘립니다.
            count_white[i + OFFSET] += 1

            # 조건 체크: 둘 다 2번 이상이면 회색(3번)
            if count_white[i + OFFSET] >= 2 and count_black[i + OFFSET] >= 2:
                tile_color[i + OFFSET] = 3
            else:
                # 아니면 현재 색깔은 흰색(1번)
                tile_color[i + OFFSET] = 1

        # 마지막으로 칠한 위치로 현재 위치를 갱신합니다.
        current_pos = start

# 4. 모든 작업이 끝난 후 색깔별로 타일 수를 셉니다.
white_ans = 0 # 최종 흰색 개수
black_ans = 0 # 최종 검은색 개수
gray_ans = 0  # 최종 회색 개수

for i in range(MAX_SIZE):
    if tile_color[i] == 1:
        white_ans += 1
    elif tile_color[i] == 2:
        black_ans += 1
    elif tile_color[i] == 3:
        gray_ans += 1

# 5. 정답 출력 (공백을 사이에 두고 출력)
print(white_ans, black_ans, gray_ans)
```

이차원 평면 위에 N개의 직사각형이 주어집니다. 이 직사각형들이 만들어내는 총 넓이를 구하는 프로그램을 작성해보세요.

```python
# 1. 입력 받기
# n: 화면에 그려질 직사각형의 개수입니다.
n = int(input())

# 2. '모눈종이' (2차원 리스트) 준비하기
# 좌표가 -100부터 100까지이므로, 가로 세로 200칸씩 필요합니다.
# OFFSET(오프셋): 음수를 양수 인덱스로 바꾸기 위한 기준값 (100)
OFFSET = 100
MAX_RANGE = 201

# 'paper' (페이퍼): 0으로 가득 찬 200x200 크기의 모눈종이를 만듭니다.
# 0이면 빈 칸, 1이면 직사각형이 채워진 칸을 의미합니다.
paper = [[0] * MAX_RANGE for _ in range(MAX_RANGE)]

# 3. N개의 직사각형을 하나씩 모눈종이에 그립니다.
for _ in range(n):
    # x1, y1 (왼쪽 아래 점), x2, y2 (오른쪽 위 점) 좌표를 받습니다.
    x1, y1, x2, y2 = map(int, input().split())

    # x축(가로) 반복문: x1부터 x2-1까지
    for x in range(x1 + OFFSET, x2 + OFFSET):
        # y축(세로) 반복문: y1부터 y2-1까지
        for y in range(y1 + OFFSET, y2 + OFFSET):
            # 해당 칸을 1로 색칠합니다.
            # 이미 1인 곳에 또 1을 써도 값은 그대로 1이므로 중복이 자동 해결됩니다!
            paper[x][y] = 1

# 4. 색칠된 칸(값이 1인 칸)의 총 개수 세기
total_area = 0 # 최종 넓이를 저장할 변수

# 모눈종이 전체를 훑으면서 1이 몇 개인지 확인합니다.
for row in range(MAX_RANGE):
    for col in range(MAX_RANGE):
        if paper[row][col] == 1:
            total_area += 1

# 5. 결과 출력
print(total_area)
```

# 겹치지 않는 사각형의 넓이

좌표평면위에 직사각형 A, B를 먼저 붙이고 그 위에 직사각형 M을 붙였습니다. 아직 남아있는 (M으로 덮이지 못한) 직사각형 A, B의 넓이의 합을 구하는 프로그램을 작성해보세요. 단, 직사각형 A, B는 겹치지 않게 주어진다고 가정해도 좋습니다.

```python
# 1. '모눈종이' (2차원 리스트) 준비하기
# 좌표 범위가 -1000 ~ 1000 이므로, 가로 세로 2000칸 정도가 필요합니다.
# OFFSET (오프셋): 음수를 0 이상의 인덱스로 바꾸기 위한 보정값입니다.
OFFSET = 1000
MAX_RANGE = 2001

# 'paper' (페이퍼): 0으로 채워진 거대한 2D 리스트를 만듭니다.
paper = [[0] * MAX_RANGE for _ in range(MAX_RANGE)]

# 2. 직사각형 A와 B 정보 입력받고 색칠하기
# A와 B는 겹치지 않는다고 했으니 순서대로 칠하면 됩니다.
for _ in range(2):
    # x1, y1 (좌측 하단), x2, y2 (우측 상단)
    x1, y1, x2, y2 = map(int, input().split())

    # x축(가로) 루프
    for x in range(x1 + OFFSET, x2 + OFFSET):
        # y축(세로) 루프
        for y in range(y1 + OFFSET, y2 + OFFSET):
            # 사각형 A, B가 있는 자리를 1로 표시합니다.
            paper[x][y] = 1

# 3. 직사각형 M 정보 입력받고 '지우기'
# M의 좌표를 입력받습니다.
mx1, my1, mx2, my2 = map(int, input().split())

# M이 차지하는 범위를 다시 0으로 덮어버립니다. (지우개 역할)
for x in range(mx1 + OFFSET, mx2 + OFFSET):
    for y in range(my1 + OFFSET, my2 + OFFSET):
        # M으로 덮이는 곳은 다시 빈 칸(0)으로 만듭니다.
        paper[x][y] = 0

# 4. 남아있는 면적(값이 1인 칸) 계산하기
remaining_area = 0 # Remaining Area (리메이닝 에어리어): 남은 넓이

# 모눈종이 전체를 돌며 1의 개수를 셉니다.
for row in range(MAX_RANGE):
    for col in range(MAX_RANGE):
        if paper[row][col] == 1:
            remaining_area += 1

# 5. 결과 출력
print(remaining_area)
```

좌표평면위에 가로세로 길이가 8이고 넓이가 64인 색종이가 N장 주어집니다. 이 N장 색종이의 각 좌측하단의 꼭지점이 주어졌을 때 모든 색종이가 붙여진 이후의 총 넓이를 구하는 프로그램을 작성해보세요. 단, 모든 색종이는 좌표평면위에서 (−100, −100)을 좌측하단으로 (100, 100)을 우측상단으로 하는 정사각형 범위를 벗어나지 않는다고 가정해도 좋습니다. 또한 겹치는 영역은 단 한번만 넓이에 포함시킵니다.

```python
# 1. 입력 받기: n은 색종이의 장수입니다.
n = int(input())

# 2. 모눈종이(2차원 리스트) 준비
# 좌표가 -100 ~ 100이므로 전체 길이는 200입니다.
# OFFSET(오프셋)을 100으로 두어 음수 좌표를 양수 인덱스(0~200)로 바꿉니다.
OFFSET = 100
MAX_SIZE = 201
# 'grid' (그리드): 격자, 모눈종이라는 뜻입니다. 전부 0으로 채워둡니다.
grid = [[0] * MAX_SIZE for _ in range(MAX_SIZE)]

# 3. 색종이 한 장씩 처리하기
for _ in range(n):
    # 좌측 하단 꼭지점 좌표 (x, y)를 입력받습니다.
    # map(int, input().split())은 "숫자들을 정수로 바꿔서 쪼개줘"라는 뜻입니다.
    x_start, y_start = map(int, input().split())

    # 색종이의 가로세로 길이가 8이므로, 끝나는 지점을 계산합니다.
    x_end = x_start + 8
    y_end = y_start + 8

    # x축(가로) 반복: 시작점부터 끝점 직전까지 8칸을 돕니다.
    for i in range(x_start + OFFSET, x_end + OFFSET):
        # y축(세로) 반복: 시작점부터 끝점 직전까지 8칸을 돕니다.
        for j in range(y_start + OFFSET, y_end + OFFSET):
            # 색종이가 덮인 칸을 1로 표시합니다.
            # 이미 1인 곳에 또 1을 써도 값은 1이므로 중복이 자동 해결됩니다!
            grid[i][j] = 1

# 4. 전체 넓이(1의 개수) 계산하기
# total_area (토탈 에어리어): 총 넓이
total_area = 0

for row in range(MAX_SIZE):
    for col in range(MAX_SIZE):
        # 만약 이 칸의 값이 1이라면 (색종이가 한 번이라도 붙었다면)
        if grid[row][col] == 1:
            total_area += 1

# 5. 정답 출력
print(total_area)
```

```python
# 1. 입력 받기
n, m = map(int, input().split())

# 2. A와 B의 위치를 매 초마다 저장할 리스트를 만듭니다.
# 0초일 때의 위치는 둘 다 0입니다.
pos_a = [0]
pos_b = [0]

# 3. A의 이동 기록하기
current_pos = 0 # current_pos (커런트 포스): 현재 위치
for _ in range(n):
    d, t = input().split() # d: 방향(direction), t: 시간(time)
    t = int(t)

    for _ in range(t):
        if d == 'R':
            current_pos += 1
        else:
            current_pos -= 1
        # 매 초마다 변하는 위치를 리스트에 추가(append)합니다.
        pos_a.append(current_pos)

# 4. B의 이동 기록하기
current_pos = 0 # B를 위해 위치를 다시 0으로 초기화합니다.
for _ in range(m):
    d, t = input().split()
    t = int(t)

    for _ in range(t):
        if d == 'R':
            current_pos += 1
        else:
            current_pos -= 1
        pos_b.append(current_pos)

# 5. 최초로 만나는 시간 찾기
# 0초(출발점)는 제외하고 1초부터 비교해야 합니다.
ans = -1 # ans (앤서): 정답, 못 만나면 -1을 출력해야 하므로 초기값을 -1로 둡니다.

# A와 B가 이동한 총 시간은 항상 동일하다고 문제에 주어졌습니다.
total_time = len(pos_a)

for i in range(1, total_time):
    # i초일 때 A의 위치와 B의 위치가 같다면 "만났다!"
    if pos_a[i] == pos_b[i]:
        ans = i
        break # 최초의 만남을 찾는 것이므로 찾자마자 반복문을 나갑니다.

# 6. 결과 출력
print(ans)
```

```python
# 1. 한 줄에 있는 n, m, k를 동시에 입력받습니다.
n, m, k = map(int, input().split())

# 2. 벌칙을 받은 학생 번호 m개를 리스트로 한 번에 저장합니다. (리스트 컴프리헨션)
# student_list (스튜던트 리스트): 벌칙 학생 번호들이 담긴 명단
student_list = [int(input()) for _ in range(m)]

# 3. 각 학생의 벌칙 횟수를 저장할 장부(리스트)를 만듭니다.
# 학생 번호 1~n을 그대로 인덱스로 쓰기 위해 n+1 크기로 만듭니다.
penalty_counts = [0] * (n + 1)

# 4. 정답을 담을 변수 (찾지 못하면 -1)
ans = -1

# 5. 명단에 적힌 학생 번호를 하나씩 꺼내서 확인합니다.
for num in student_list:
    # 해당 학생 번호의 벌칙 횟수를 1 늘립니다.
    penalty_counts[num] += 1

    # 벌칙 횟수가 기준 k에 도달했는지 확인합니다.
    if penalty_counts[num] >= k:
        ans = num # 이 학생이 첫 벌금 대상자!
        break     # 찾았으므로 즉시 멈춤

# 6. 최종 정답 출력
print(ans)
```

```python
# 1. 입력 받기
n, m = map(int, input().split())

# 2. A와 B의 위치를 초 단위로 기록할 리스트
# 출발점(0초)의 위치는 0입니다.
pos_a = [0]
pos_b = [0]

# 3. 로봇 A의 이동 기록
current_pos = 0
for _ in range(n):
    t, d = input().split() # t: 시간, d: 방향
    for _ in range(int(t)):
        if d == 'R': current_pos += 1
        else: current_pos -= 1
        pos_a.append(current_pos)

# 4. 로봇 B의 이동 기록
current_pos = 0
for _ in range(m):
    t, d = input().split()
    for _ in range(int(t)):
        if d == 'R': current_pos += 1
        else: current_pos -= 1
        pos_b.append(current_pos)

# 5. [중요] 두 리스트의 길이를 똑같이 맞추기
# 더 짧게 움직인 로봇은 마지막 위치에 계속 머물러야 합니다.
max_time = max(len(pos_a), len(pos_b))

# A가 더 빨리 멈췄다면, 남은 시간 동안 마지막 위치를 채워줍니다.
while len(pos_a) < max_time:
    pos_a.append(pos_a[-1]) # pos_a[-1]은 리스트의 마지막 원소를 뜻합니다.

# B가 더 빨리 멈췄다면 마찬가지로 채워줍니다.
while len(pos_b) < max_time:
    pos_b.append(pos_b[-1])

# 6. "직전엔 다르고 지금은 같은" 경우 세기
count = 0
# 0초는 시작점이니 제외하고 1초부터 비교합니다.
for i in range(1, max_time):
    # 조건 1: i-1초(직전)에는 서로 다른 위치였는가?
    # 조건 2: i초(현재)에는 서로 같은 위치인가?
    if pos_a[i-1] != pos_b[i-1] and pos_a[i] == pos_b[i]:
        count += 1

# 7. 결과 출력
print(count)
```

단, 전염된 사람끼리 만나도 전염시킨 것으로 간주해야 합니다. 예를 들어 전염된 x 개발자와 전염된 y 개발자끼리 악수를 해도 전염병을 옮기게 되는 횟수에 포함시켜야 합니다. 이때, 감염 횟수에 포함될 뿐, 이미 전염되었던 사람이 재감염이 되는 것은 아님에 유의합니다.

```python
# 1. 입력 받기
# n: 개발자 수, k: 전염 가능 횟수, p: 최초 감염자, t: 악수 횟수
n, k, p, t = map(int, input().split())

# 악수 정보를 리스트 컴프리헨션으로 한 번에 받기
# [시간, x, y] 형태로 저장됩니다.
handshakes = [list(map(int, input().split())) for _ in range(t)]

# 2. 시간 순서(첫 번째 요소)대로 정렬하기
# 파이썬의 sort()는 리스트의 첫 번째 값을 기준으로 오름차순 정렬해줍니다.
handshakes.sort()

# 3. 개발자들의 상태를 기록할 장부 만들기
# 인덱스 편의를 위해 n+1 크기로 만듭니다.
# 'infected' (인펙티드): 감염 여부 (0: 건강, 1: 감염)
infected = [0] * (n + 1)
infected[p] = 1 # 최초 감염자 p를 표시합니다.

# 'transfer_count' (트랜스퍼 카운트): 병을 옮긴 횟수를 기록합니다.
transfer_count = [0] * (n + 1)

# 4. 시간 순서대로 악수를 진행합니다.
for time, x, y in handshakes:
    # 나중에 상태가 변할 수 있으므로, 현재의 감염 상태를 미리 저장해둡니다.
    # (동시에 악수할 때 서로 영향을 주는 것을 정확히 처리하기 위함)
    x_is_infected = infected[x]
    y_is_infected = infected[y]

    # 케이스 1: x가 감염자이고, 아직 k번 미만으로 옮겼을 때
    if x_is_infected and transfer_count[x] < k:
        # y를 감염시키고 (이미 걸렸어도 상관없음), x의 횟수를 늘립니다.
        infected[y] = 1
        transfer_count[x] += 1

    # 케이스 2: y가 감염자이고, 아직 k번 미만으로 옮겼을 때
    # (if가 아니라 elif를 쓰면 안 됩니다! 둘 다 감염자일 수도 있기 때문이죠.)
    if y_is_infected and transfer_count[y] < k:
        # x를 감염시키고, y의 횟수를 늘립니다.
        infected[x] = 1
        transfer_count[y] += 1

# 5. 결과 출력
# 1번 개발자부터 n번 개발자까지 감염 여부를 공백 없이 출력합니다.
for i in range(1, n + 1):
    print(infected[i], end="")
```

```python
# 1. 입력 받기
n = int(input())
people = list(map(int, input().split()))

# 2. 최솟값 상자 만들기 (직접 큰 숫자 넣기)
# 문제의 범위를 아득히 초과하는 큰 값을 넣어두면 됩니다.
min_total_distance = 10**9 # 10의 9제곱 (10억)

# 3. 모든 집(i)을 모임 장소로 가정해보기 (완전 탐색)
for i in range(n):
    current_dist_sum = 0 # 이번 모임(i)의 총 거리 합계

    # 4. 각 집(j)에 사는 사람들이 i번 집까지 오는 거리 계산
    for j in range(n):
        # abs(i-j)는 두 집 사이의 거리, people[j]는 사람 수
        current_dist_sum += abs(i - j) * people[j]

    # 5. 지금까지의 최솟값보다 지금 결과가 더 작으면 갱신!
    if current_dist_sum < min_total_distance:
        min_total_distance = current_dist_sum

# 6. 정답 출력
print(min_total_distance)
```

```python
# 1. 입력 받기
n = int(input())
# 2차원 격자 정보를 리스트 컴프리헨션으로 한 줄씩 받습니다.
grid = [list(map(int, input().split())) for _ in range(n)]

# 3. 최댓값을 저장할 변수 (동전의 개수는 0개 이상이므로 0으로 시작)
max_coins = 0

# 4. 모든 행(i)에 대해 탐색 시작
for i in range(n):
    # 각 행에서 가로로 3칸을 잡아야 하므로 j의 범위를 조절합니다.
    # j+2가 마지막 인덱스(n-1)를 넘지 않으려면 j는 n-3까지 가야 합니다.
    # range(n - 2)는 n-3까지 숫자를 만들어줍니다! (예: n=5면 0, 1, 2까지)
    for j in range(n - 2):
        # 현재 위치(j)부터 연속된 3칸의 동전 개수를 합산합니다.
        current_sum = grid[i][j] + grid[i][j + 1] + grid[i][j + 2]

        # 지금까지 구한 최댓값과 비교하여 더 큰 값을 저장합니다.
        if current_sum > max_coins:
            max_coins = current_sum

# 5. 최종 결과 출력
print(max_coins)
```

이동은 항상 점프를 통해서만 가능합니다. 또, 점프 진행시 항상 현재 위치에 적혀있는 색과, 점프한 이후의 칸에 적혀있는 색이 달라야만 합니다.

점프 진행시 현재 위치에서 적어도 한칸 이상 오른쪽에 있는 위치이며 동시에 현재 위치에서 적어도 한칸 이상 아래쪽에 있는 위치인 곳으로만 점프가 가능합니다.

정확히 시작, 도착 지점을 제외하고 점프하며 도달한 위치가 정확히 2곳 뿐이어야 합니다.

```python
# 1. 입력 받기 (리스트 컴프리헨션 활용)
r, c = map(int, input().split())
# input().split()을 써서 공백이 섞여 들어와도 안전하게 리스트로 만듭니다.
grid = [input().split() for _ in range(r)]

count = 0

# 2. 첫 번째 경유지 (r1, c1) 선택
# (0, 0)보다는 크고, 적어도 (r-2, c-2)보다는 작거나 같아야 함
for r1 in range(1, r):
    for c1 in range(1, c):

        # 3. 두 번째 경유지 (r2, c2) 선택
        # 첫 번째 경유지보다 아래/오른쪽이어야 하고,
        # 동시에 도착점(r-1, c-1)보다 한 칸 위/왼쪽에서 끝나야 함 (그래서 range의 끝이 r-1)
        for r2 in range(r1 + 1, r - 1):
            for c2 in range(c1 + 1, c - 1):

                # 4. 모든 점프 조건(색깔이 달라야 함) 확인
                # 시작점 -> 1번 지점 색깔 체크
                cond1 = (grid[0][0] != grid[r1][c1])
                # 1번 지점 -> 2번 지점 색깔 체크
                cond2 = (grid[r1][c1] != grid[r2][c2])
                # 2번 지점 -> 도착점 색깔 체크
                cond3 = (grid[r2][c2] != grid[r-1][c-1])

                # 세 가지 조건을 모두 만족하면 정답!
                if cond1 and cond2 and cond3:
                    count += 1

# 5. 결과 출력
print(count)
```

```python
# 1. 입력 받기
n = int(input())
points = [list(map(int, input().split())) for _ in range(n)]

# 2. 아주 큰 숫자를 직접 입력 (sys 안 씀!)
min_total_distance = 10**9 # 10의 9제곱 (10억)
# sys 없이 무한대 설정하기
# min_total_distance = float('inf')

# 3. 하나씩 건너뛰며 계산하기
for skip_idx in range(1, n - 1):
    current_dist = 0
    prev_idx = 0

    for i in range(1, n):
        if i == skip_idx:
            continue

        # 택시 거리 계산
        dist = abs(points[i][0] - points[prev_idx][0]) + \
               abs(points[i][1] - points[prev_idx][1])

        current_dist += dist
        prev_idx = i

    # 최솟값 비교
    if current_dist < min_total_distance:
        min_total_distance = current_dist

# 4. 결과 출력
print(min_total_distance)
```

```python
# 1. 입력 받기
n, k = map(int, input().split())

# 2. 1차원 직선(배열) 준비 (위치 범위가 0~100이므로 크기는 101)
# 낭떠러지 계산을 편하게 하기 위해 앞뒤로 여유를 줘도 좋지만,
# 여기서는 인덱스 체크를 직접 해보겠습니다.
line = [0] * 101

for _ in range(n):
    amount, pos = map(int, input().split())
    # 같은 위치에 여러 바구니가 가능하므로 += 사용
    line[pos] += amount

max_candies = 0

# 3. 모든 가능한 중심점 c를 다 해봅니다.
for c in range(101):
    current_sum = 0

    # 4. 중심 c로부터 거리 k 이내인 범위를 계산합니다. [c-k, c+k]
    # 시작점(start)은 0보다 작아질 수 없고, 끝점(end)은 100보다 커질 수 없습니다.
    start = max(0, c - k)
    end = min(100, c + k)

    # 해당 범위의 사탕을 다 더합니다.
    # sum(line[start : end + 1]) -> end까지 포함해야 하므로 +1!
    current_sum = sum(line[start : end + 1])

    # 5. 최댓값 갱신
    if current_sum > max_candies:
        max_candies = current_sum

# 6. 결과 출력
print(max_candies)
```

```python
# 첫 줄에 N
N = int(input())
a, b, c = map(int, input().split())

# 2. 모든 조합 만들기 (3중 반복문)
cnt = 0
for i in range(1, N + 1):         # 첫 번째 자리에 1~N까지 다 넣어봄
    for j in range(1, N + 1):     # 두 번째 자리에 1~N까지 다 넣어봄
        for k in range(1, N + 1): # 세 번째 자리에 1~N까지 다 넣어봄

            # 조건 검사
            # abs(a - i)는 "정답 a"와 "내가 뽑은 i"의 차이(거리)입니다.
            # <= 2는 그 거리가 2 이내(0, 1, 2)인지 확인하는 겁니다.
            if abs(a - i) <= 2 or abs(b - j) <= 2 or abs(c - k) <= 2:
                # 위 3개 중 하나라도(or) 참이면 카운트합니다.
                cnt += 1

print(cnt)
```

```python
# 1. 입력 받기
abilities = list(map(int, input().split()))
n = 6
total_sum = sum(abilities)

# 2. 최솟값을 찾기 위해 아주 큰 숫자로 시작
min_diff = float('inf')

# 3. 6명 중 3명을 뽑는 삼중 루프
# 중복해서 뽑지 않도록 i < j < k 순서로 범위를 잡습니다.
for i in range(n):
    for j in range(i + 1, n):
        for k in range(j + 1, n):

            # A팀의 능력치 합계
            sum_a = abilities[i] + abilities[j] + abilities[k]

            # B팀의 능력치 합계 (전체 - A팀)
            sum_b = total_sum - sum_a

            # 두 팀의 점수 차이 (절대값)
            diff = abs(sum_a - sum_b)

            # 최솟값 갱신
            if diff < min_diff:
                min_diff = diff

# 4. 결과 출력
print(min_diff)
```

# 숫자 카운트

50XP

평균 129분

66% 정답률

총 제출 1,383회
___
A와 B는 서로 숫자 맞추는 게임을 하려고 합니다.

A가 1에서 9까지의 서로 다른 숫자 세 개로 구성된 세 자리 수를 마음속으로 생각합니다.
B는 1에서 9까지의 서로 다른 숫자 세 개로 구성된 세 자리 수를 A에게 묻습니다.
B가 말한 세 자리 수에 있는 숫자들 중 하나가 A가 생각한 세 자리 수의 동일한 자리에 위치한다면 1번 카운트가 하나 올라갑니다.
B가 말한 세 자리 수에 있는 숫자들 중 하나가 A가 생각한 세 자리 수에 있긴 하나 다른 자리에 위치하면 2번 카운트가 하나 올라갑니다.
예를 들어, A가 123을 생각했고

B가 345를 물었다면 1번 카운트는 0, 2번 카운트는 1입니다.

B가 183을 물었다면 1번 카운트는 2, 2번 카운트는 0입니다.

B가 327을 물었다면 1번 카운트는 1, 2번 카운트는 1입니다.

A는 B가 말한 수에 대한 카운트를 답해주고 있습니다. B가 물어본 N개의 수에 대한 카운트 정보가 각각 주어졌을 때, B가 유추할 수 있는 수가 총 몇 개인지 알아맞히는 프로그램을 작성해보세요.

```python
# 1. 입력 받기
n = int(input())
queries = []
for _ in range(n):
    # [숫자(문자열), 1번카운트, 2번카운트] 형태로 저장
    num, c1, c2 = input().split()
    queries.append([num, int(c1), int(c2)])

ans_count = 0

# 2. 모든 가능한 서로 다른 3자리 수 생성 (1~9)
for i in range(1, 10):
    for j in range(1, 10):
        for k in range(1, 10):
            # 조건: 서로 다른 숫자여야 함
            if i == j or j == k or i == k:
                continue

            # 현재 우리가 '정답'이라고 가정한 수
            candidate = [i, j, k]
            is_possible = True

            # 3. B가 했던 모든 질문(queries)과 대조해보기
            for q_num, target_c1, target_c2 in queries:
                c1 = 0 # 1번 카운트 (숫자&자리 일치)
                c2 = 0 # 2번 카운트 (숫자만 일치)

                # 질문받은 숫자도 리스트로 변환 (예: '123' -> [1, 2, 3])
                q_list = [int(q_num[0]), int(q_num[1]), int(q_num[2])]

                for idx1 in range(3):
                    for idx2 in range(3):
                        # 숫자가 일단 존재하면
                        if candidate[idx1] == q_list[idx2]:
                            # 자리까지 같으면 1번 카운트
                            if idx1 == idx2:
                                c1 += 1
                            # 자리가 다르면 2번 카운트
                            else:
                                c2 += 1

                # 하나라도 카운트가 다르면 이 후보는 가짜!
                if c1 != target_c1 or c2 != target_c2:
                    is_possible = False
                    break

            # 4. 모든 질문을 통과했다면 정답 후보로 인정
            if is_possible:
                ans_count += 1

# 5. 결과 출력
print(ans_count)
```

```python
# 1. 입력 받기
abilities = list(map(int, input().split()))
n = 6
total_sum = sum(abilities)

min_diff = float('inf')

# 2. 첫 번째 팀 (i, j) 고르기
for i in range(n):
    for j in range(i + 1, n):

        # 3. 두 번째 팀 (k, l) 고르기
        for k in range(n):
            for l in range(k + 1, n):
                # 중요: 이미 1팀에 뽑힌 사람(i, j)은 2팀에 뽑힐 수 없음!
                if k == i or k == j or l == i or l == j:
                    continue

                # 팀별 합계 계산
                sum1 = abilities[i] + abilities[j]
                sum2 = abilities[k] + abilities[l]
                sum3 = total_sum - sum1 - sum2

                # 세 팀 중 최대값과 최소값 찾기
                max_team = max(sum1, sum2, sum3)
                min_team = min(sum1, sum2, sum3)

                # 그 차이가 역대급으로 작다면 갱신!
                diff = max_team - min_team
                if diff < min_diff:
                    min_diff = diff

# 4. 결과 출력
print(min_diff)
```

```python
# 1. 입력 받기
n, b = map(int, input().split())
prices = [int(input()) for _ in range(n)]

max_students = 0

# 2. 어떤 학생의 선물을 할인받을지 하나씩 정해봅니다 (discount_idx)
for discount_idx in range(n):
    # 이번 차례에 할인받은 가격들을 담을 리스트
    temp_prices = []

    for i in range(n):
        if i == discount_idx:
            # 선택된 학생은 반값!
            temp_prices.append(prices[i] // 2)
        else:
            temp_prices.append(prices[i])

    # 3. 선물을 최대한 많이 주려면 '싼 가격'부터 골라야 합니다 (정렬)
    temp_prices.sort()

    # 4. 예산(b) 안에서 몇 명까지 살 수 있는지 계산
    current_count = 0
    current_sum = 0

    for p in temp_prices:
        if current_sum + p <= b:
            current_sum += p
            current_count += 1
        else:
            # 예산 초과면 더 이상 못 삼
            break

    # 최댓값 갱신
    if current_count > max_students:
        max_students = current_count

# 5. 결과 출력
print(max_students)
```

```python
# 1. 입력 받기
n, m, d, s = map(int, input().split())

# 먹은 기록: (사람, 치즈, 시간)
eat_records = []
for _ in range(d):
    eat_records.append(list(map(int, input().split())))

# 아픈 기록: (사람, 시간)
sick_records = []
for _ in range(s):
    sick_records.append(list(map(int, input().split())))

max_medicine = 0

# 2. 1번 치즈부터 m번 치즈까지 하나씩 범인이라고 가정해봅니다.
for target_cheese in range(1, m + 1):
    # 이 치즈가 범인일 가능성이 있는지 체크
    is_possible = True

    for sick_p, sick_t in sick_records:
        # 아픈 사람(sick_p)이 이 치즈(target_cheese)를 먹었었는지 확인
        ate = False
        for p, m_idx, t in eat_records:
            # 아프기 시작한 시간(sick_t)보다 최소 1초 전(t < sick_t)에 먹었어야 함
            if p == sick_p and m_idx == target_cheese and t < sick_t:
                ate = True
                break

        # 아픈 사람 중 한 명이라도 이 치즈를 먹은 적이 없다면 범인이 아님!
        if not ate:
            is_possible = False
            break

    # 3. 범인일 가능성이 있다면, 이 치즈를 먹은 모든 사람 수를 셉니다.
    if is_possible:
        medicine_count = 0
        # 중복을 피하기 위해 사람별로 체크
        for person in range(1, n + 1):
            for p, m_idx, t in eat_records:
                if p == person and m_idx == target_cheese:
                    medicine_count += 1
                    break # 한 번이라도 먹었으면 약 1개 필요

        # 4. 최대 약 개수 갱신
        if medicine_count > max_medicine:
            max_medicine = medicine_count

# 5. 결과 출력
print(max_medicine)
```

```python
# 1. 입력 받기
n = int(input())
seats = list(input()) # 수정 가능하도록 리스트로 변환

# 가장 가까운 두 사람 간의 거리의 최댓값을 저장할 변수
max_min_dist = 0

# 2. 모든 좌석을 하나씩 확인하며 빈자리에 앉혀봅니다.
for i in range(n):
    if seats[i] == '0':
        # 일단 앉혀봅니다.
        seats[i] = '1'

        # 3. 현재 배치에서 '가장 가까운 두 사람 간의 거리'를 구합니다.
        # 사람들의 위치(인덱스)만 따로 뽑아냅니다.
        pos = []
        for j in range(n):
            if seats[j] == '1':
                pos.append(j)

        # 인접한 사람들 사이의 거리 중 최솟값을 찾습니다.
        min_dist = n # 충분히 큰 값으로 초기화
        for k in range(len(pos) - 1):
            dist = pos[k+1] - pos[k]
            if dist < min_dist:
                min_dist = dist

        # 4. 이 '최솟값'들 중에서 가장 큰 값을 갱신합니다.
        if min_dist > max_min_dist:
            max_min_dist = min_dist

        # 다음 시도를 위해 다시 빈자리로 되돌려 놓습니다. (중요!)
        seats[i] = '0'

# 5. 결과 출력
print(max_min_dist)
```

```python
import sys

# 1. 입력 받기
n = int(input())
seats = list(input())

max_min_dist = 0

# 2. 첫 번째 사람이 앉을 자리 i를 고릅니다.
for i in range(n):
    if seats[i] == '0':
        seats[i] = '1' # 첫 번째 사람 착석

        # 3. 두 번째 사람이 앉을 자리 j를 고릅니다. (i보다 뒤에 있는 자리만 보면 중복 방지!)
        for j in range(i + 1, n):
            if seats[j] == '0':
                seats[j] = '1' # 두 번째 사람 착석

                # 4. 현재 배치에서 가장 가까운 두 사람 간의 거리 계산
                pos = [idx for idx, val in enumerate(seats) if val == '1']

                min_dist = n
                for k in range(len(pos) - 1):
                    dist = pos[k+1] - pos[k]
                    if dist < min_dist:
                        min_dist = dist

                # 최댓값 갱신
                if min_dist > max_min_dist:
                    max_min_dist = min_dist

                seats[j] = '0' # 두 번째 사람 원상복구

        seats[i] = '0' # 첫 번째 사람 원상복구

# 5. 결과 출력
print(max_min_dist)
```

```python
# 1. 입력 받기
n = int(input())
sums = list(map(int, input().split()))

found = False # 정답을 찾았는지 표시하는 깃발

# 2. 첫 번째 숫자(first) 가정
for first in range(1, n + 1):
    if found: break # 이미 정답을 찾았다면 더 이상 돌지 않음

    a = [0] * n
    a[0] = first
    used = [False] * (n + 1)
    used[first] = True

    is_possible = True
    for i in range(n - 1):
        next_val = sums[i] - a[i]

        if next_val < 1 or next_val > n or used[next_val]:
            is_possible = False
            break

        a[i + 1] = next_val
        used[next_val] = True

    if is_possible:
        # 정답 출력
        print(*(a))
        found = True # 깃발을 올림
```

코디는 적절한 위치에 그물을 던져 최대한 많은 수의 물고기를 잡고 싶어 합니다.

바다의 정보와 그물의 크기 K가 주어질 때, 코디가 잡을 수 있는 물고기의 최대 마릿수를 계산하는 프로그램을 작성하세요.

```python
# 1. 입력 받기
n, k = map(int, input().split())
sea = [list(map(int, input().split())) for _ in range(n)]

max_fish = 0
possible = False # 그물을 한 번이라도 던질 수 있었는지 체크

# 2. 그물을 놓을 수 있는 시작점 (r, c) 탐색
for r in range(n - k + 1):
    for c in range(n - k + 1):

        current_fish = 0
        has_obstacle = False

        # [1] 위쪽 변 & 아래쪽 변 한꺼번에 체크 (가로줄 두 개)
        for j in range(k):
            # 위쪽 변 (r행)
            if sea[r][c + j] == -1:
                has_obstacle = True; break
            current_fish += sea[r][c + j]

            # 아래쪽 변 (r + k - 1행)
            if sea[r + k - 1][c + j] == -1:
                has_obstacle = True; break
            current_fish += sea[r + k - 1][c + j]

        if has_obstacle: continue # 암초 있으면 다음 시작점으로

        # [2] 왼쪽 변 & 오른쪽 변 체크 (세로줄 두 개, 꼭짓점 제외)
        # 위에서 양 끝(꼭짓점)을 더했으니 i는 1부터 k-2까지만!
        for i in range(1, k - 1):
            # 왼쪽 변 (c열)
            if sea[r + i][c] == -1:
                has_obstacle = True; break
            current_fish += sea[r + i][c]

            # 오른쪽 변 (c + k - 1열)
            if sea[r + i][c + k - 1] == -1:
                has_obstacle = True; break
            current_fish += sea[r + i][c + k - 1]

        if has_obstacle: continue

        # 3. 암초가 없었다면 최댓값 갱신
        if not possible:
            max_fish = current_fish
            possible = True
        else:
            if current_fish > max_fish:
                max_fish = current_fish

# 4. 결과 출력
print(max_fish if possible else 0)
```

# 카드 섞기

N장의 카드가 쌓여 있는 카드 더미가 있습니다 (N은 짝수). 코디는 카드들을 아래의 규칙에 따라 새롭게 섞습니다.

1단계: 더미 나누기
전체 카드 더미를 위에서부터 순서대로 절반씩 나누어 왼쪽 더미와 오른쪽 더미를 만듭니다. 각 더미에는 N/2장의 카드가 들어있으며, 기존의 순서가 유지됩니다.

2단계: 카드 선택 및 재배열
두 더미에 모두 카드가 남아있다면, 다음 과정을 반복하여 새로운 카드 더미를 만듭니다.

카드 선택: 왼쪽 더미의 맨 위 카드와 오른쪽 더미의 맨 위 카드에 적힌 수의 합을 확인합니다.
합이 홀수이면: 왼쪽 더미의 맨 위 카드를 뽑습니다.
합이 짝수이면: 오른쪽 더미의 맨 위 카드를 뽑습니다.
남은 카드 처리: 한쪽 더미가 모두 비게 되면, 남은 더미에 있는 카드들을 위에서부터 순서대로 모두 가져와 새로운 더미 뒤에 붙입니다.
3단계: 결과
카드를 뽑은 순서대로 쌓은 것이 최종적으로 섞인 카드 더미가 됩니다.

```python
# 1. 입력 받기
n = int(input())
cards = list(map(int, input().split()))

# 1단계: 더미 나누기
half = n // 2
left = cards[:half]
right = cards[half:]

# 2단계: 카드 선택 및 재배열
# i는 왼쪽 더미의 손가락, j는 오른쪽 더미의 손가락입니다.
i = 0
j = 0
result = []

# 두 손가락이 모두 더미 범위 안에 있는 동안 반복
while i < half and j < half:
    # 현재 손가락이 가리키는 카드들의 합 확인
    card_sum = left[i] + right[j]

    if card_sum % 2 != 0: # 합이 홀수이면
        result.append(left[i]) # 결과에 담고
        i += 1                 # 왼쪽 손가락만 한 칸 옆으로!
    else: # 합이 짝수이면
        result.append(right[j]) # 결과에 담고
        j += 1                  # 오른쪽 손가락만 한 칸 옆으로!

# 3단계: 남은 카드 처리
# i나 j가 가리키는 곳부터 끝까지 남은 카드들을 싹 붙여줍니다.
result.extend(left[i:])
result.extend(right[j:])

# 결과 출력
print(*(result))
```

코디는 숙면을 취하기 위해 모든 전등을 꺼진 상태로 만들고자 합니다.

초기 전등의 상태가 주어졌을 때, 모든 전등을 꺼진 상태로 만들기 위한 최소 스위치 동작 횟수를 출력하는 프로그램을 작성하세요.

전망이 가장 좋은 집의 전망 수치를 출력하는 프로그램을 작성하세요.

```python
# 1. 입력 받기
n = int(input())
h = list(map(int, input().split()))

max_view_score = 0

# 2. 각 집(i)을 기준으로 전망 수치를 계산합니다.
for i in range(n):
    current_height = h[i]
    current_score = 0

    # 3. 내 왼쪽(j = i-1)부터 0번 집까지 거꾸로 가며 확인합니다.
    for j in range(i - 1, -1, -1):
        # 만약 왼쪽 집이 나보다 높거나 같으면 탐색 중단!
        if h[j] >= current_height:
            break

        # 나보다 낮은 집이라면 높이 차이를 점수에 더합니다.
        current_score += abs(current_height - h[j])

    # 4. 가장 좋은 전망 수치를 저장합니다.
    if current_score > max_view_score:
        max_view_score = current_score

# 5. 최종 결과 출력
print(max_view_score)
```

# 회전 퍼즐

정사각형 격자 위에 마름모 형태로 구성된 회전 퍼즐이 존재합니다. 퍼즐은 중심칸을 1번부터 N번까지 총 N개의 구역이 감싸고 있는 형태이며 각 구역에는 정확히 하나의 뚫린 칸이 존재합니다. 뚫린 칸은 중심칸을 기준으로 상하좌우, 한 방향만을 선택해 도달할 수 있는 칸들중 하나입니다.

Imagec249ef73-e68f-41c1-ad7b-29643481eae4.webp

이 퍼즐은 특정 구역을 정해 시계 방향으로 90도 회전시킬 수 있습니다. 90도 회전된 구역은 다음과 같이 뚫린 칸의 위치가 바뀝니다.

Image864d940b-b1e5-46dd-913d-7999d6ce0ad6.webp

단, i번 구역을 회전시키면 그보다 바깥쪽에 있는 모든 구역, i+1,i+2,…,N번 구역이 함께 회전합니다.

퍼즐을 풀기 위해선 모든 구역의 뚫린 칸이 하나의 방향을 향하도록 정렬해야 합니다.

주어진 퍼즐 상태에서, 모든 구역의 뚫린 칸을 일렬로 맞추기 위한 최소 회전 횟수를 출력하는 프로그램을 작성하세요.

```python
n = int(input())
s = input().strip()

# 방향 순서: 위(U) -> 오른쪽(R) -> 아래(D) -> 왼쪽(L) (시계 방향 90도씩)
directions = "URDL"

def get_min_rotations(target_dir):
    total_rotations = 0
    current_accumulated_rot = 0 # 안쪽 구역들 때문에 누적된 회전 수

    for i in range(n):
        # 1. 현재 구역의 처음 방향 인덱스 찾기
        start_idx = directions.find(s[i])

        # 2. 안쪽에서 누적된 회전 반영 (안쪽이 돈 만큼 나도 돌아가 있음)
        # (처음 위치 + 누적 회전) % 4
        current_idx = (start_idx + current_accumulated_rot) % 4

        # 3. 목표 방향(target_dir)까지 시계방향으로 몇 번 더 돌려야 하는가?
        target_idx = directions.find(target_dir)
        needed_rot = (target_idx - current_idx + 4) % 4

        # 4. 돌려야 한다면 횟수 추가 및 누적치 갱신
        total_rotations += needed_rot
        current_accumulated_rot += needed_rot

    return total_rotations

# 상, 하, 좌, 우 네 방향 각각으로 맞췄을 때의 횟수 중 최솟값 찾기
results = []
for char in "URDL":
    results.append(get_min_rotations(char))

print(min(results))
```
